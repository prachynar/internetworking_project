```python
import pandas as pd
import numpy as np
import os

# path (same as yours)
path = "/kaggle/input/network-intrusion-dataset"

# load all CSVs
files = os.listdir(path)
dfs = []

for file in files:
    if file.endswith(".csv"):
        df = pd.read_csv(f"{path}/{file}")
        dfs.append(df)

# combine
data = pd.concat(dfs, ignore_index=True)

# clean column names
data.columns = data.columns.str.strip()

# remove duplicates
data.drop_duplicates(inplace=True)

# convert label to binary
data['Label'] = data['Label'].apply(
    lambda x: 'Benign' if x == 'BENIGN' else 'Malicious'
)

# fix inf and NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(0, inplace=True)

print("Data ready:", data.shape)
```

```python
selected_features = [
    'Packet Length Mean',
    'Packet Length Std',
    'Average Packet Size',
    'Flow Duration',
    'Flow Bytes/s',
    'Flow Packets/s',
    'Flow IAT Mean',
    'Flow IAT Std',
    'Flow IAT Min',
    'Flow IAT Max',
    'Total Fwd Packets',
    'Total Backward Packets',
    'Total Length of Fwd Packets',
    'Total Length of Bwd Packets',
    'Fwd Packet Length Mean',
    'Bwd Packet Length Mean'
]
```

```python
plt.figure(figsize=(20, 8))
sns.boxplot(data=data[selected_features], orient='h')
plt.title('Boxplots of Selected Behavioural Features')
plt.xlabel('Feature Value')
plt.show()
```
